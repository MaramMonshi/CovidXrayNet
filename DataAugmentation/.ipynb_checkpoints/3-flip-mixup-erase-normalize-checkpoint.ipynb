# Data Augmentation

## Flipping, Mixup, Random Erasing and Normalizing CXRs

```python
from fastai.vision.all import *
path = Path('/home/jupyter/covidcxr') 
```

```python
torch.cuda.empty_cache()
```

```python
# Fix result
def seed_everything(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.cuda.manual_seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False
SEED = 42
seed_everything(SEED)
```

```python
df = pd.read_csv(path/'covidcxr.csv')
df
```

```python
get_x=lambda x:path/f"{x[0]}"
get_y=lambda x:x[1]
splitter=RandomSplitter(seed=SEED)
bs=64
```

```python
arch = resnet18
metrics=[accuracy,  
         RocAuc(average='macro', multi_class='ovr'),
         MatthewsCorrCoef(sample_weight=None),
         Precision(average='macro'),
         Recall(average='macro'),     
         F1Score(average='macro')] 
```

### Flip

```python
item_tfms=Resize(480, method='squish', pad_mode='zeros', resamples=(2, 0))
batch_tfms=[*aug_transforms(mult=1.0, do_flip=True, flip_vert=True, 
                            max_rotate=0.0, max_zoom=0.0, max_lighting=0.0, max_warp=0.0, 
                            p_affine=0.75, p_lighting=0.75, 
                            xtra_tfms=None, size=None, mode='bilinear', pad_mode='reflection', 
                            align_corners=True, batch=False, min_scale=1.0)]
```

```python
db = DataBlock(blocks=(ImageBlock(cls=PILImageBW), CategoryBlock),
               get_x=get_x,
               get_y=get_y,
               splitter=splitter,
               item_tfms = item_tfms,
               batch_tfms=batch_tfms)
```

```python
dl = db.dataloaders(df, bs=bs)
```

```python
dl.show_batch(max_n=6, nrows=1, ncols=6)
```

```python
learn = cnn_learner(dl, arch=arch, metrics=metrics)
```

```python
learn.fine_tune(30)
```

### Mixup

```python
item_tfms=Resize(480, method='squish', pad_mode='zeros', resamples=(2, 0))
batch_tfms=[*aug_transforms(mult=1.0, do_flip=False, flip_vert=False, 
                            max_rotate=0.0, max_zoom=0.0, max_lighting=0.0, max_warp=0.0, 
                            p_affine=0.75, p_lighting=0.75, 
                            xtra_tfms=None, size=None, mode=None, pad_mode=None, 
                            align_corners=True, batch=False, min_scale=1.0)]
```

```python
db = DataBlock(blocks=(ImageBlock(cls=PILImageBW), CategoryBlock),
               get_x=get_x,
               get_y=get_y,
               splitter=splitter,
               item_tfms = item_tfms,
               batch_tfms=batch_tfms)
```

```python
dl = db.dataloaders(df, bs=bs)
```

```python
mixup = MixUp(0.4)
learn = cnn_learner(dl, arch=arch, cbs=mixup, metrics=metrics)
```

```python
dl.show_batch(max_n=6, nrows=1, ncols=6)
```

```python
learn.fine_tune(30)
```

### Random Erase

```python
additional_aug = [RandomErasing(p=1., sl=0.1, sh=0.0, min_aspect=0.3, max_count=6)] 
```

```python
item_tfms=Resize(480, method='squish', pad_mode='zeros', resamples=(2, 0))
batch_tfms=[*aug_transforms(mult=1.0, do_flip=False, flip_vert=False, 
                            max_rotate=0.0, max_zoom=0.0, max_lighting=0.0, max_warp=0.0, 
                            p_affine=0.75, p_lighting=0.75, 
                            xtra_tfms=additional_aug, size=None, mode=None, pad_mode=None, 
                            align_corners=True, batch=False, min_scale=1.0)]
```

```python
db = DataBlock(blocks=(ImageBlock(cls=PILImageBW), CategoryBlock),
               get_x=get_x,
               get_y=get_y,
               splitter=splitter,
               item_tfms = item_tfms,
               batch_tfms=batch_tfms)
```

```python
dl = db.dataloaders(df, bs=bs)
```

```python
dl.show_batch(max_n=6, nrows=1, ncols=6)
```

```python
learn = cnn_learner(dl, arch=arch, metrics=metrics)
```

```python
learn.fine_tune(30)
```

### Normalize

```python
item_tfms=Resize(480, method='squish', pad_mode='zeros', resamples=(2, 0))
batch_tfms=Normalize.from_stats(*imagenet_stats)
```

```python
db = DataBlock(blocks=(ImageBlock(cls=PILImageBW), CategoryBlock),
               get_x=get_x,
               get_y=get_y,
               splitter=splitter,
               item_tfms = item_tfms,
               batch_tfms=batch_tfms)
```

```python
dl = db.dataloaders(df, bs=bs)
```

```python
dl.show_batch(max_n=6, nrows=1, ncols=6)
```

```python
learn = cnn_learner(dl, arch=arch, metrics=metrics)
```

```python
learn.fine_tune(30)
```
